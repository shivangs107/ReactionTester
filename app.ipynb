```python
import pygame
import random
import time

pygame.init()

# Window setup
WIDTH, HEIGHT = 700, 500
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Reaction Time Tester")

font = pygame.font.Font(None, 40)

clock = pygame.time.Clock()
running = True

# Game states
waiting_for_green = True
reaction_time = None
start_time = 0
green_time = 0

# Function to draw centered text
def draw_text(text, y, color=(255,255,255)):
    label = font.render(text, True, color)
    rect = label.get_rect(center=(WIDTH//2, y))
    screen.blit(label, rect)

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        # When user presses SPACE
        if event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
            # If reacted too early (before green)
            if waiting_for_green:
                reaction_time = "Too Early! Wait for GREEN."
            else:
                reaction_time = round((time.time() - green_time) * 1000)  # In ms
                waiting_for_green = True

    # Clear screen
    screen.fill((0,0,0))

    # If waiting for green light
    if waiting_for_green:
        draw_text("Press SPACE to Start", HEIGHT//2 - 80)
        draw_text("Wait for the screen to turn GREEN...", HEIGHT//2 - 30)

        # After pressing SPACE, begin random wait
        if reaction_time != None:
            if isinstance(reaction_time, str):
                draw_text(reaction_time, HEIGHT//2 + 40, (255,100,100))
            else:
                draw_text(f"Reaction Time: {reaction_time} ms", HEIGHT//2 + 40, (100,255,100))

        # SPACE pressed → start random delay
        keys = pygame.key.get_pressed()
        if keys[pygame.K_SPACE]:
            wait_time = random.uniform(1.5, 3.0)  # random 1.5–3 seconds
            start_time = time.time()
            waiting_for_green = False

    # If waiting to show green
    else:
        # If still in waiting period → red screen
        if time.time() - start_time < wait_time:
            screen.fill((150,0,0))
            draw_text("WAIT...", HEIGHT//2, (255,255,255))
        else:
            # Turn screen GREEN
            screen.fill((0,150,0))
            draw_text("GO!", HEIGHT//2 - 50, (255,255,255))
            draw_text("Press SPACE Now!", HEIGHT//2, (255,255,255))

            # Mark the time when screen turned green
            if green_time == 0:
                green_time = time.time()

    # Reset green time after reaction
    if waiting_for_green and reaction_time != None:
        green_time = 0

    pygame.display.flip()
    clock.tick(60)

pygame.quit()
```
